**8.Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted
from any YouTube Video.**

```python
driver = webdriver.Chrome()
video = input("Enter the youtube video link:  ")
driver.get(video)

comment = []
comment_time = []
comment_upvote = []

for _ in range(20):
    driver.execute_script("window.scrollBy(0,1000)")
    
    try:
        comment_tags = driver.find_elements(By.CLASS_NAME,'style-scope yt-formatted-string')
        for i in comment_tags[0:500]:
            comment.append(i.text.split('\n')[2])
            comment_time.append(i.text.split('\n')[1])
            comment_upvote.append(i.text.split('\n')[-2])
    except NoSuchElementException:
        comment.append('-')
        comment_time.append('-')
        comment_upvote.append('-')


print(len(comment ), len(comment_time), len(comment_upvote))



df = pd.DataFrame({'comment': comment, 'comment_time': comment_time, 'comment_upvote': comment_upvote})
df
```

**9.Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in
“London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall
reviews, privates from price, dorms from price, facilities and property description.**

```python
driver = webdriver.Chrome()
driver.get("http://www.hostelworld.com/")

try:
    # Click on the location input field and enter "London"
    loc = driver.find_element(By.XPATH, '//input[@id="home-search-keywords"]')
    loc.click()
    loc.send_keys('London')

    # Select the second option from the predicted search results
    option = driver.find_element(By.XPATH, '//li[@id="predicted-search-results-0"]')
    option.click()

    # Click on the search button
    search_button = driver.find_element(By.XPATH, '//button[@id="search-button"]')
    search_button.click()
except NoSuchElementException:
    pass

try:
    # Find all the hostel names and links
    hostel_elements = driver.find_elements(By.XPATH, '//h2[@class="title title-6"]/a')
    for element in hostel_elements:
        name.append(element.text)
        link.append(element.get_attribute("href"))
except NoSuchElementException:
    pass

try:
    # Find all the hostel distances
    distance_elements = driver.find_elements(By.CLASS_NAME, "description")
    for element in distance_elements:
        distance.append(element.text)
except NoSuchElementException:
    pass

try:
    # Find all the hostel ratings
    rating_elements = driver.find_elements(By.CLASS_NAME, "score")
    for element in rating_elements:
        rating.append(element.text)
except NoSuchElementException:
    pass

try:
    # Find all the hostel reviews
    review_elements = driver.find_elements(By.CLASS_NAME, "reviews")
    for element in review_elements:
        review.append(element.text)
except NoSuchElementException:
    pass

try:
    # Find all the hostel prices
    price_elements = driver.find_elements(By.CLASS_NAME, "price-col")
    for element in price_elements:
        prices.append(element.text)
except NoSuchElementException:
    pass

# Split prices into private and dorm
private = prices[::2]
dorm = prices[1::2]

try:
    # Find all the hostel descriptions
    description_elements = driver.find_elements(By.CLASS_NAME, "info")
    for element in description_elements[1:]:
        des.append(element.text.split('\n')[:3])
except NoSuchElementException:
    pass

try:
    # Find all the hostel facilities
    facility_elements = driver.find_elements(By.CLASS_NAME, "facilities-label.facilities")
    for element in facility_elements:
        fac.append(element.text)
except NoSuchElementException:
    pass

# Create a DataFrame
data = {
    'Name': name,
    'Distance': distance,
    'Rating': rating,
    'Review': review,
    'Private Price': private,
    'Dorm Price': dorm,
    'Description': des,
    'Facilities': fac,
    'Link': link
}

df = pd.DataFrame(data)

print(df)
```
