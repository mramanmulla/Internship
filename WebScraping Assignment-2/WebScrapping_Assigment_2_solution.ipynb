Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data. This task will be done in following steps:
1. First get the webpage https://www.shine.com/
2. Enter “Data Analyst” in “Job title, Skills” field and enter “Bangalore” in “enter the  location” field.
3. Then click the search button.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data. Note: All of the above steps have to be done in code. No step is to be done manually.

```python
page = requests.get('https://www.shine.com/job-search/data-analyst-jobs-in-bangalore?q=Data%20Analyst&loc=Bangalore')

page

shine_page = BeautifulSoup(page.content)

shine_page

# will scrap 1st title
header = []

for i in shine_page.find_all ('h2', itemprop='name'):
    header.append(i.text)

header

# will scrap 2nd title
posting_time = []

for i in shine_page.find_all ('div',class_="jobCard_jobCard_features__wJid6"):
    posting_time.append(i.text)

posting_time

# will scrap 3rd title

type = []

for i in shine_page.find_all ('ul',class_="jobCard_jobCard_jobDetail__jD82J"):
    type.append(i.text)

type

# will scrap 4th title

company_name = []

for i in shine_page.find_all ('div',class_="jobCard_jobCard_cName__mYnow"):
    company_name.append(i.text)

company_name

# will check lenth for each title

print(len(header))
print(len(posting_time))
print(len(type))
print(len(company_name))

# Make data frame of top 10 data analyst job as per site
data_analyst=pd.DataFrame({})
data_analyst['header']=header[:10]
data_analyst['posting_time']=posting_time[:10]
data_analyst['type']=type[:10]
data_analyst['company_name']=company_name[:10]
data_analyst
```

Q8: Write a python program to scrape data for Top 1000 Quotes of All Time. The above task will be done in following steps:
1. First get the webpagehttps://www.azquotes.com/
2. Click on Top Quotes
3. Than scrap a) Quote b) Author c) Type Of Quotes



```python
page = requests.get('https://www.azquotes.com/top_quotes.html')

page

azquotes = BeautifulSoup(page.content)

azquotes

# Will scrap for Quote

Quote = []

for i in azquotes.find_all ('a',class_="title"):
    Quote.append(i.text)

Quote


# Will scrap Author

Author = []

for i in azquotes.find_all ('div',class_="author"):
    Author.append(i.text.replace('\n',''))

Author

# Will scrap Type Of Quotes

TypeOfQuotes = []

for i in azquotes.find_all ('div',class_="tags"):
    TypeOfQuotes.append(i.text.replace('\n',''))

TypeOfQuotes


azquotes_df=pd.DataFrame({})
azquotes_df['Quote']=Quote
azquotes_df['Author']=Author
azquotes_df['TypeOfQuotes']=TypeOfQuotes
azquotes_df
```

Q10: Write a python program to display list of 50 Most expensive cars in the world (i.e. Car name and Price) from https://www.motor1.com/

2. Then You have to type in the search bar ’50 most expensive cars’
3. Then click on 50 most expensive cars in the world..

```python
page = requests.get('https://www.motor1.com/features/308149/most-expensive-new-cars-ever/')

page

expensive_car = BeautifulSoup(page.content)

expensive_car


# will scrap car name

car_name = []

for i in expensive_car.find_all ('h3',class_="subheader"):
    car_name.append(i.text)

car_name

# will scrap price

car_price = []

for i in expensive_car.find_all ('strong'):
    car_price.append(i.text.replace('Price:',''))

car_price

top50cars=pd.DataFrame({})
top50cars['car_name']=car_name[:50]
top50cars['car_price']=car_price[:50]

top50cars
```
